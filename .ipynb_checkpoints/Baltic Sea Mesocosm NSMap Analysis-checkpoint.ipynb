### Effect of Time Window on Baltic Sea Species

```python
# Calanoids
ts = standardize(table[:,1])
plotTS(ts)
```

```python
year_steps = 52
```

```python
maxLen = ts.shape[0]
lengths = np.arange(208,maxLen+1,step=year_steps)

E_max = 4
tau = 3

# length, starting index, delta, theta, r_sqrd
n_rows = 78
results = np.zeros((n_rows, 5))
tally = 0

for length in lengths:
    for start in np.arange(0, maxLen - length+1, step=year_steps):
        ts_chunk = ts[start:length+start]
        delta, theta, r_sqrd = get_delta_agg(ts_chunk, E_max, tau=tau, return_forecast_skill=True)
        results[tally] = np.array([length, start, delta, theta, r_sqrd])
        tally += 1
        # print(tally)
```

```python
# np.savetxt("calanoids_time_window.csv",results, fmt="%i,%i,%0.5f,%0.5f,%0.5f")
results = np.loadtxt("calanoids_time_window.csv", delimiter=",")
```

```python
nonstationarityHeatMap = np.zeros((12,12)) * np.nan
for l, o, d, t, r in results:
    i_length = 11 - int(l / year_steps - 4)
    i_offset = int(o / year_steps)
    nonstationarityHeatMap[i_length, i_offset] = d
```

```python
fig, ax = plt.subplots(1,figsize=(13,10))
# im = ax.imshow(nonstationarityHeatMap[3:,:-3])
im = ax.imshow(nonstationarityHeatMap[:-3], cmap='summer')
# ax.set_xlabel("Start year",fontsize=15)
# ax.set_xticks(np.arange(29,step=2))
# ax.set_xticks([])
ax.set_xticks(np.arange(0,22+1,step=1))
ax.set_xlim(-0.55,11.44)
# ax.set_xticklabels(np.arange(1996,2023,step=2))

ax.set_ylabel("Time series length (years)",fontsize=15)
# ax.set_yticks(np.arange(21+1))
# ax.set_yticklabels(np.arange(21+1)+5)
# ax.set_ylim(2,25)
# ax.set_yticks(np.arange(-1,22+1))
ax.set_yticklabels(np.flip(np.hstack([np.arange(4,22+1,step=2),""])))

divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
cb = plt.colorbar(im, cax=cax)
cb.set_label(r"$\overline{\delta}$",fontsize=15)

axTS = divider.append_axes('bottom', size='30%', pad=0.1)

axTS.plot(np.linspace(0, 22.05,num=len(ts)), ts,c="green")
axTS.set_xticks(np.arange(0,22,step=2))
axTS.set_yticks([])
axTS.set_ylabel("Calanoids",fontsize=15)
axTS.set_xlabel("Start year",fontsize=15)

plt.show()
```

```python
for i, species in zip([1,2,4], ["calanoids","rotifers","picophytoplankton"]):
    ts = standardize(table[:,i])
    maxLen = ts.shape[0]
    year_steps = 26
    lengths = np.arange(52,maxLen+1,step=year_steps)

    E_max = 4
    tau = 3

    # length, starting index, delta, theta, r_sqrd
    n_rows = 435
    results = np.zeros((n_rows, 5))
    tally = 0

    for length in lengths:
        for start in np.arange(0, maxLen - length+1, step=year_steps):
            
            ts_chunk = ts[start:length+start]
            delta, theta, r_sqrd = get_delta_agg(ts_chunk, E_max, tau=tau, return_forecast_skill=True)
            results[tally] = np.array([length, start, delta, theta, r_sqrd])
            
            tally += 1
            
    np.savetxt(f"{species}_time_window.csv",results, fmt="%i,%i,%0.5f,%0.5f,%0.5f")
```

```python
for index in range(3):
    species_list = ["calanoids","rotifers","nanophytoplankton","picophytoplankton"]
    results = np.loadtxt(f"{species_list[index]}_time_window.csv", delimiter=",")
    
    ts = standardize(table[:,index])
    nonstationarityHeatMap = makeHeatMapBaltic(results)
    print(nonstationarityHeatMap)
    makeFigureBaltic(nonstationarityHeatMap, ts)
```

```python
def makeHeatMapBaltic(results):
    nonstationarityHeatMap = np.zeros((29,29)) * np.nan
    for l, o, d, t, r in results:
        i_length = 28-(int(l / year_steps) - 2)
        i_offset = int(o / year_steps)
        # print(i_length, i_offset)
        nonstationarityHeatMap[i_length, i_offset] = d
        
    return nonstationarityHeatMap
```

```python
def makeFigureBaltic(nonstationarityHeatMap, ts):
    fig, ax = plt.subplots(1,figsize=(13,10))
    # im = ax.imshow(nonstationarityHeatMap[3:,:-3])
    im = ax.imshow(nonstationarityHeatMap, cmap='summer')
    # ax.set_xlabel("Start year",fontsize=15)
    # ax.set_xticks(np.arange(29,step=2))
    # ax.set_xticks([])
    # ax.set_xticks(np.arange(0,22+1,step=1))
    ax.set_xlim(-1.3,32)
    # ax.set_xticklabels(np.arange(1996,2023,step=2))

    ax.set_ylabel("Time series length (years)",fontsize=15)
    # yticks = np.arange(0,29,step=2)
    #ax.set_ylim([yticks[-1]+0.5,yticks[0]-0.5])
    #ax.set_yticks(yticks)
    # ax.set_yticklabels(np.flip(np.hstack([year_steps * yticks,""])))

    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cb = plt.colorbar(im, cax=cax)
    cb.set_label(r"$\overline{\delta}$",fontsize=15)

    axTS = divider.append_axes('bottom', size='30%', pad=0.2)

    axTS.plot(np.arange(len(ts)) / year_steps, ts,c="green")
    # axTS.set_xticks(np.arange(0,24,step=2))
    # axTS.set_yticks([])
    axTS.set_ylabel("Nanophytoplankton",fontsize=15)
    axTS.set_xlabel("Start year",fontsize=15)

    plt.show()
```
