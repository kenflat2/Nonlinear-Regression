```python
import numpy as np
import numpy.random as rand
import matplotlib.pyplot as plt
```

```python
N = 700# number of timesteps
p = 1 # points in ball in initial coniditons
ballRadius = 0.1 # radius of init cond ball
ballCenter = 0.5 

X = np.zeros((N,p))
X[0] = np.linspace(ballCenter-ballRadius,ballCenter+ballRadius,p)# rand.normal(ballCenter, ballRadius, (p))

r = 3

for i in range(N-1):
    X[i+1] = r * X[i] * (1 - X[i])
    
    if i == int(N/2):
        X[i+1] += 0.1
```

```python
figLog, axLog = plt.subplots(1,figsize=(12,6))

for pi in range(p):
    axLog.plot(X[:,pi], c="r",alpha=0.1)
    
plt.show()
```
